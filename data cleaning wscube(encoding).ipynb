### the conversion of data which is in the categorical data to the numerical format so that it would be better for model to understand mainly three types they are onehot encodeing label encoding and the ordinal encoding

```python
# Categorical Data Without Order
# Example: ['Apple', 'Banana', 'Orange'] → [0, 1, 2]
# Used in: Gender classification, binary categories, etc.
```

# ordinal encoding

```python
# Ordinal encoding is mainly used when: Categorical Data Has an Order
# Example: ['Low', 'Medium', 'High'] → [0, 1, 2]
# Used in: Customer ratings, education levels, etc.
```

```python
df = pd.DataFrame({"Size" : ["s","m","l","xl","s","s","m","l","xl","s","s"]})
## here for ordinal data encoding we have to define the order for it 
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = ord_data)
df["en_data"] =  oe.fit_transform(df[['Size']])
```

```python
# we can also do this with the mapping 
# we have to define the order first
ord_data1 = {"s":1,"m":2,"l":3,"xl":4}
df["encoded_map"] = df["Size"].map(ord_data1)
```
