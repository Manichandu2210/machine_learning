```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"mulreg.csv")
```

```python
sns.pairplot(data = df)
plt.show()
```

```python
df.rename(columns = {"income":"salary"},inplace = True)
```

```python
sns.heatmap(data = df.corr(),annot = True)
```

```python
x= df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
lr.score(xtest,ytest)*100
```

```python
lr.predict(xtest)
```

```python
# how this value is going to predict like the linear regression it will predict in such a way that 
# y = m1x1+m2x2+c
lr.coef_
```

```python
lr.intercept_
```

```python
# so if we give the value to lr.predict([[25,1]]) it will predict in such a way that 
m1 = lr.coef_[0]
m2 = lr.coef_[1]
x1,x2 = 25,1
c = lr.intercept_
y = m1*x1+m2*x2+c
y
```
